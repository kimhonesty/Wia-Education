# 02. 그래프 축(Axis) 설정
---
### 그래프 축 커스터마이징
- 그래프의 가독성을 높이기 위해 축(axis)에 대한 다양한 설정이 필요하다.
- X축과 Y축에 라벨을 추가하거나, 눈금을 조정하여 데이터를 더 명확하게 표현할 수 있다.


```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

```python
x = [1,2,3]
y = [2,4,8]
```

## 제목에 폰트 스타일 적용
- title() 함수의 fontdict 파라미터를 사용하여 제목의 폰트와 크기를 변경할 수 있다.


```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family' :'Malgun Gothic', 'size' : 25})
plt.show()
```

## 축 라벨 추가
### xlabel과 ylabel
- **xlabel()**: X축에 라벨을 추가한다.
- **ylabel()**: Y축에 라벨을 추가한다.
- 축 라벨을 추가하면 그래프가 무엇을 나타내는지 더 명확하게 전달할 수 있다.


```python
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('Y축')

plt.title('꺾은선 그래프', fontdict={'family' :'Malgun Gothic', 'size' : 25})
plt.show()
```

### 축 라벨 색상 설정
- color 파라미터를 사용하여 축 라벨의 색상을 변경할 수 있다.
- 색상은 이름(예: 'red') 또는 RGB 코드(예: '#00aa00')로 지정 가능하다.


```python
plt.plot(x,y)
plt.xlabel('X축', color = 'red')
plt.ylabel('Y축', color = '#00aa00')

plt.title('꺾은선 그래프', fontdict={'family' :'Malgun Gothic', 'size' : 25})
plt.show()
```

### 축 라벨 위치 설정
- **loc** 파라미터를 사용하여 라벨의 위치를 조정할 수 있다.
- xlabel: 'left', 'center', 'right'
- ylabel: 'bottom', 'center', 'top'


```python
plt.plot(x,y)

plt.xlabel('X축', color = 'red', loc='right')
plt.ylabel('Y축', color = '#00aa00', loc='top')

plt.title('꺾은선 그래프', fontdict={'family' :'Malgun Gothic', 'size' : 25})
plt.show()
```

## 축 눈금 설정
### xticks와 yticks
- **xticks()**: X축의 눈금 위치를 지정한다.
- **yticks()**: Y축의 눈금 위치를 지정한다.
- 원하는 위치에만 눈금을 표시하여 그래프를 더 깔끔하게 만들 수 있다.


```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])

plt.show()
```

### 축 눈금 라벨 변경
- xticks와 yticks 함수에 두 번째 인자로 라벨 리스트를 전달하면 눈금 라벨을 변경할 수 있다.


### 축 눈금 회전
- rotation 파라미터를 사용하여 눈금 라벨을 회전시킬 수 있다.
- 긴 라벨이 겹칠 때 유용하다.


```python
plt.plot(x,y)
plt.xticks([1,2,3], ['첫째', '둘째', '셋째'])
plt.yticks([2,4,8], ['낮음', '중간', '높음'])

plt.show()
```

## 축 범위 설정
### xlim과 ylim
- **xlim()**: X축의 표시 범위를 설정한다.
- **ylim()**: Y축의 표시 범위를 설정한다.
- 데이터의 특정 구간을 확대하거나 축소하여 보고 싶을 때 사용한다.


```python
plt.plot(x,y)
plt.xlim(0,4)
plt.ylim(0,10)
plt.show()
```

### 축 눈금 회전


```python
labels = ['January', 'Febuary', 'March']
values = [10,20,15]

plt.plot(labels, values)
plt.xticks(rotation=45)
plt.show()
```
