인구 피라미드


```python
import pandas as pd

df_m = pd.read_excel(
    "data/2509_인구현황.xlsx", skiprows=3,
    index_col='행정기관', usecols='B, E:Y')

print(df_m.dtypes)

df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)

df_m.head(3)
```

```python
df_w = pd.read_excel(
    "data/2509_인구현황.xlsx", skiprows=3,
    index_col='행정기관', usecols='B, AB:AV')

print(df_w.dtypes)

df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)

df_w.head(3)

df_w.columns = df_m.columns

df_w
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

```python
plt.barh(df_m.columns, df_m.iloc[0] // 1000) ## 단위 : 천명
plt.barh(df_w.columns, -df_w.iloc[0] // 1000) ## 단위 : 천명
plt.title('2025년 9월 인구 피라미드')


plt.show()
```

```python
age_groups = df_m.columns   # 연령대
men = df_m.iloc[0] // 1000 # 남자 천명 단위
women = df_w.iloc[0] // 1000 # 여자 천명 단위

plt.figure(figsize=(10, 6))                 # 그래프 크기 조절

# 남자: 왼쪽으로, 여자: 오른쪽으로 (막대 색상 다르게)
plt.barh(age_groups, men, color='steelblue', label='남자')
plt.barh(age_groups, -women, color='lightcoral', label='여자')

# 축 설정: x축 라벨, 범위, 눈금 설정
plt.xlabel('인구수 (천명)', fontsize=13)
plt.ylabel('연령대', fontsize=13)
plt.xlim(-max(women)*1.1, max(men)*1.1)              # 여유 있게 축

plt.title('2025년 9월 인구 피라미드', fontsize=15)

# x축에 남자/여자 방향 설명
plt.xticks(
    list(range(int(-max(women)*1.1), int(max(men)*1.1)+1, 10)),
    [abs(x) for x in range(int(-max(women)*1.1), int(max(men)*1.1)+1, 10)]
)
plt.text(max(men)*0.7, len(age_groups)-2, '남자', color='steelblue', fontsize=13)
plt.text(-max(women)*0.8, len(age_groups)-2, '여자', color='lightcoral', fontsize=13)

# 중심선
plt.axvline(0, color='gray', linestyle='--')

# 범례
plt.legend(loc='upper right')

# y축 뒤집기 (최연소가 아래로)
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```
